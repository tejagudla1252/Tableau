```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load the data
data = pd.read_csv("Agrofood_co2_emission.csv")
print(data.columns)
missing_values = data.isnull().sum()
print(missing_values)
print(data.count)
```

```python
data.dropna(subset=["Crop Residues","Forestland","Net Forest conversion","Food Household Consumption","IPPU","Manure applied to Soils","Manure Management","On-farm energy use"],axis = 0,inplace = True)
```

```python
data.isnull().sum()
```

```python
# Plot 1: Line Plot - Emissions Trends Over the Years
plt.figure(figsize=(10, 6))
plt.plot(data['Year'], data['total_emission'])
plt.xlabel('Year')
plt.ylabel('Total Emission')
plt.title('Emissions Trends Over the Years')
plt.grid(True)
plt.show()
```

```python
# Plot 2: Bar Plot - Comparison of Emissions from Different Sources
plt.figure(figsize=(10, 6))
sources = ['Savanna fires', 'Forest fires', 'Crop Residues', 'Rice Cultivation', 'Drained organic soils (CO2)']
emissions = data[sources].sum()
plt.bar(sources, emissions)
plt.xlabel('Emission Sources')
plt.ylabel('Total Emission')
plt.title('Comparison of Emissions from Different Sources')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot 3: Scatter Plot - Relationship Between Temperature and Emissions
plt.figure(figsize=(10, 6))
plt.scatter(data['Average Temperature °C'], data['total_emission'])
plt.xlabel('Average Temperature °C')
plt.ylabel('Total Emission')
plt.title('Relationship Between Temperature and Emissions')
plt.grid(True)
plt.show()
```

```python
# Plot 4: Stacked Bar Plot - Population Distribution in Urban and Rural Areas
plt.figure(figsize=(10, 6))
urban_pop = data['Urban population']
rural_pop = data['Rural population']
plt.bar(data['Area'], urban_pop, label='Urban Population')
plt.bar(data['Area'], rural_pop, bottom=urban_pop, label='Rural Population')
plt.xlabel('Area')
plt.ylabel('Population')
plt.title('Population Distribution in Urban and Rural Areas')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot 5: Area Plot - Contribution of Different Sources to Total Emissions
plt.figure(figsize=(10, 6))
sources = ['Fires in organic soils', 'Fires in humid tropical forests', 'On-farm energy use']
emissions = data[sources].sum(axis=1)
plt.fill_between(data['Year'], emissions, color='skyblue')
plt.xlabel('Year')
plt.ylabel('Total Emission')
plt.title('Contribution of Different Sources to Total Emissions')
plt.grid(True)
plt.show()
```

```python
# Plot 6: Box Plot - Distribution of Average Temperature
plt.figure(figsize=(8, 6))
plt.boxplot(data['Average Temperature °C'])
plt.ylabel('Average Temperature °C')
plt.title('Distribution of Average Temperature')
plt.grid(True)
plt.show()
```

```python
# Write the modified dataframe back to a CSV file
data.to_csv('Agrofood_co2_emission_cleaned.csv', index=False)
```
