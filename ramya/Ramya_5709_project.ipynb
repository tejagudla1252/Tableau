```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('Latest_Data_Science_Salaries.csv')
print(df.head())
```

```python
print(df.isnull().sum())
```

```python
# 1: Bar Plot - Top Job Titles with Highest Average Salary
avg_salary_by_job_title = df.groupby('Job Title')['Salary in USD'].mean().sort_values(ascending=False).head(10)
plt.bar(avg_salary_by_job_title.index, avg_salary_by_job_title.values)
plt.xlabel('Job Title')
plt.ylabel('Average Salary (USD)')
plt.title('Top 10 Job Titles with Highest Average Salary')
plt.xticks(rotation=90)
plt.show()
```

```python
# 2: Box Plot - Salary Distribution by Experience Level
plt.figure(figsize=(8, 6))  # Set the figure size to avoid overlapping boxes
experience_order = ['Entry', 'Mid', 'Senior', 'Executive']
plt.boxplot([df[df['Experience Level'] == 'Entry']['Salary in USD'],
             df[df['Experience Level'] == 'Mid']['Salary in USD'],
             df[df['Experience Level'] == 'Senior']['Salary in USD'],
             df[df['Experience Level'] == 'Executive']['Salary in USD']],
            labels=experience_order)
plt.xlabel('Experience Level')
plt.ylabel('Salary (USD)')
plt.title('Salary Distribution by Experience Level')
plt.show()
```

```python
# 3: Line Plot - Average Salary by Company Location
avg_salary_by_company_location = df.groupby('Company Location')['Salary in USD'].mean().sort_values(ascending=False)
plt.plot(avg_salary_by_company_location.index, avg_salary_by_company_location.values, marker='o', linestyle='-')
plt.xlabel('Company Location')
plt.ylabel('Average Salary (USD)')
plt.title('Average Salary by Company Location')
plt.xticks(rotation=90)
plt.show()
```

```python
# Example 4: Box Plot - Salary Distribution by Experience Level
plt.boxplot(df.groupby('Experience Level')['Salary in USD'].median())
plt.xlabel('Experience Level')
plt.ylabel('Median Salary (USD)')
plt.title('Salary Distribution by Experience Level')
plt.show()
```

```python
# 5: Pie Chart - Distribution of Employment Types
employment_types = df['Employment Type'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(employment_types, labels=employment_types.index, autopct='%1.1f%%', startangle=140)
plt.legend(title='Employment Types', bbox_to_anchor=(1, 0.5), loc='center left')
plt.axis('equal')
plt.title('Distribution of Employment Types')
plt.show()
```
