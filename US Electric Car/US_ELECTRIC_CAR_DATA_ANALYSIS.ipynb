```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load the data
data = pd.read_csv("us_car_data.csv")
print(data.columns)
missing_values = data.isnull().sum()
print(missing_values)
print(data.count)
```

```python
# Plot 1: Distribution of model years
plt.hist(data['Model Year'])
plt.xlabel('Model Year')
plt.ylabel('Count')
plt.title('Distribution of Model Years')
plt.show()
```

```python
# Plot 2: Top 10 Cities with the highest number of electric vehicles
top_10_cities = data['City'].value_counts().head(10)

plt.plot(top_10_cities.index, top_10_cities.values)
plt.xlabel('Cities')
plt.ylabel('Number of Electric Vehicles')
plt.title('Top 10 Cities with the Highest Number of Electric Vehicles')
plt.show()
```

```python
# Plot 3: Average electric range by City
average_range_by_City = data.groupby('City')['Electric Range'].mean()

plt.bar(average_range_by_City.index, average_range_by_City.values)
plt.xlabel('City')
plt.ylabel('Average Electric Range (Miles)')
plt.title('Average Electric Range by City')
plt.show()
```

```python
# Plot 4: Correlation between base MSRP and electric range
plt.scatter(data['Base MSRP'], data['Electric Range'])
plt.xlabel('Base MSRP (USD)')
plt.ylabel('Electric Range (Miles)')
plt.title('Correlation between Base MSRP and Electric Range')
plt.show()
```

```python
# Plot 5: Distribution of CAFV (Clean Aternative Fuel Vehicle) types
cafv = data['CAFV'].value_counts()
explode = (0.1,0,0)
plt.pie(cafv.values, explode, labels=cafv.index, autopct='%1.1f%%')
plt.title('Distribution of CAFV')
plt.show()
```

```python
# Plot 6: Top 10 electric utilities with the highest number of electric vehicles
top_10_utilities = data['Electric Utility'].value_counts().head(10)

plt.bar(top_10_utilities.index, top_10_utilities.values)
plt.xlabel('Electric Utility')
plt.ylabel('Number of Electric Vehicles')
plt.title('Top 10 Electric Utilities with the Highest Number of Electric Vehicles')
plt.show()
```
